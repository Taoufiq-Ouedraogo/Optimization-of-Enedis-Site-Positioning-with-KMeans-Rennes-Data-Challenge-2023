# Traitement du fichier des communes

# Traitement du fichier des niveaux d'interventions

```python
niv_df = pd.read_csv('../data/sujet-1/niveau_interventions.csv', delimiter=';', encoding='latin-1')
niv_df.head()
print(niv_df.shape)

niv_df.columns = ['code', 'nivC', 'nivR']
niv_df.set_index(['code'], inplace=True)
niv_df.head()
```

## Etude des niveaux d'intervention

Ici on a procd a un encodage des variables qualitatives pour faciliter le traitement des donnes.
On a galement combiner les 2 niveaud d'intervention pour avoir un indicateur plus gnral.

```python
def encoding_niv(x):
    return {'Trs Bas': 1,
            'Bas': 2,
            'Moyen': 3,
            'Haut': 4,
            'Trs Haut': 5}[x]


def couleur_niv(x):
    return {1: 'b',
            2: 'g',
            3: 'y',
            4: 'orange',
            5: 'r'}[x]
```

```python
niv_df['nivC'] = niv_df['nivC'].apply(encoding_niv)
niv_df['nivR'] = niv_df['nivR'].apply(encoding_niv)
niv_df['nivIntervention'] = niv_df.sum(axis=1) / 2
niv_df['nivIntervention'] = niv_df['nivIntervention'].astype({'nivIntervention': int}) 

niv_df['color'] = niv_df['nivIntervention'].apply(couleur_niv)

niv_df.head()
```

# Modeling

```python
plt.scatter(dfw_geo.latitude, dfw_geo.longitude,color=dfw_geo.color)
```

```python
X = np.array(dfw_geo[['latitude', 'longitude']])
kmeans = KMeans(n_clusters=20, n_init='auto')
kmeans.fit(X)
```

```python
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r')
```

```python
a = dff.groupby('Depart').mean()[['Duree']]
weight2 = [d for d in a['Duree']]
while len(weight2) != dfw_geo['nivIntervention'].shape[0]:
    weight2.append(np.array(a).mean())
weight2 = np.array(weight2)
```

```python
kmeans = KMeans(n_clusters=27, n_init='auto')
weights = np.array(weight2,dfw_geo['nivIntervention'])
kmeans.fit(X, sample_weight=weight2)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r')

labels = kmeans.labels_
print(f' score calinski = {metrics.calinski_harabasz_score(X, labels)}')
print(" score silouhette = ", metrics.silhouette_score(X, labels,metric='euclidean'))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'n_clusters': range(20,30,2), 'algorithm': ['lloyd', 'elkan'], 'init': ['k-means++', 'random']}

grid_search = GridSearchCV(kmeans, param_grid, cv=5, n_jobs=-1)

grid_search.fit(X, sample_weight=weights)

print('GridSearchcy best params: ', grid_search.best_params_)

kmeans = grid_search.best_estimator_
kmeans.fit(X, sample_weight=weights)

plt.scatter(X[:,0], X[:,1], c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], c='r')

labels = kmeans.labels_
print(f' score calinski = {metrics.calinski_harabasz_score(X, labels)}')
print(" score silouhette = ", metrics.silhouette_score(X, labels,metric='euclidean'))
```

# Evaluation du Modle & Nombre de cluster

### Indice de Calinski-Harabasz
mesure la relation entre la variance intra-cluster et la variance inter-clusters
Il varie entre 0 et +infini, o une valeur plus grande indique un meilleur clustering

### Indice de silhouette
mesure la similarit entre chaque point de cluster et ses points les plus proches dans d'autres clusters
Il varie entre 0 et +infini, o une valeur plus petite indique un meilleur clustering

```python
ch_scores, silhouette_scores = [], []
for k in range(20,30):
    km = KMeans(n_clusters=k, n_init='auto', algorithm='lloyd')
    km.fit(X, sample_weight=weights)
    labels = km.labels_
    ch_scores.append(metrics.calinski_harabasz_score(X, labels))
    silhouette_scores.append(metrics.silhouette_score(X, labels, metric='euclidean'))

    
plt.figure(1,figsize=(5,3))
sns.lineplot(x=range(20,30), y=ch_scores)
plt.title("Indice de Calinski-Harabasz")

plt.figure(2,figsize=(5,3))
sns.lineplot(x=range(20,30), y=silhouette_scores)
plt.title("Indice de silouhette")
```

```python
def nb_BO(model, df0, threshold=4):
    dfCopie = df0
    dfCopie['clust'] = model.labels_
    moy_clust = dfCopie[['nivIntervention', 'clust']].groupby('clust').agg('mean')
    moy_clust = moy_clust.sort_values(by='nivIntervention', ascending=False)
    maxx = max(moy_clust.nivIntervention) 
    maxx = "{:.1f}".format(maxx)
    print(f'n = {model.n_clusters},  max niv = {maxx} ')
    return max(moy_clust.nivIntervention)

for k in range(20,30):
    km = KMeans(n_clusters=k, n_init='auto', algorithm='lloyd')
    km.fit(X, sample_weight=weights)
    nb_BO(km, dfw_geo)
```

```python
coul = ["darkorange", "gold", "lawngreen", "lightseagreen", "green", "blue", "orange", "pink", "yellow",
        "silver", "tan", "wheat", "teal", "lime", "seagreen", "darkviolet", "darkslateblue", "royalblue",
        "indigo", "forestgreen", "darkviolet", "bisque", "gainsboro", "black", "maroon", "darkred", 
        "rosybrown", "gray", "grey","lightgray"]


key = range(0, dfw_geo.clust.max()+1)
val = coul[:dfw_geo.clust.max()+1]


    
def couleur_final(x):
    d = {}
    for k,v in zip(key,val):
        d[k] = v
    return d[x]
```

```python
dfw_geo['coulFinal'] = dfw_geo['clust'].apply(couleur_final)
dfw_geo.loc[:1205,'geometry'].plot(figsize=(10,10), color=dfw_geo['coulFinal'])
```
